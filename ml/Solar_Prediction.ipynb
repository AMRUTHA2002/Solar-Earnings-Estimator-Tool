Hence, we tried building a model to estimate the cost and power generated by solar power in a particular region by weather forecast and motivate to adopt these methods for a better living.

This dataset computes the solar power generation units based on the various geographical factors and climate of a region. It depends on 16 attributes:

- Day of year

- Day
- First hour of period

- Average Temperature (Day)
- Average wind direction (Day)
- Average wind speed (Day)
- sky cover
- visibility
- Relative humidity
- Average wind speed (Period)

- power generated

The dataset was separated into X and y, with X being my prediction data and y being what is intended to predict. For this, they are divided into training (70%) and testing (30%).

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

Scaling data is an important preprocessing step in machine learning, especially for algorithms that are sensitive to the scale of the input features. The primary reason for scaling data before performing a train-test split is to ensure that the machine learning algorithm works effectively and produces accurate results.

```python
from sklearn.preprocessing import StandardScaler
from sklearnex import patch_sklearn
patch_sklearn()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
from sklearnex.linear_model import LinearRegression
lr_algo = LinearRegression()
lr_algo.fit(X_train_scaled,y_train)
coef = lr_algo.coef_

intercept = lr_algo.intercept_
print("The accuracy for Linear Regression is : ",lr_algo.score(X_test_scaled,y_test)*100,"%")
```

```python
from sklearnex.ensemble import RandomForestRegressor
rfc_algo = RandomForestRegressor(n_estimators=100,max_depth=5)
rfc_algo.fit(X_train_scaled,y_train)
predicted_labels = rfc_algo.predict(X_test_scaled)
print("The accuracy for Random forest Regression is : ",rfc_algo.score(X_test_scaled,y_test)*100,"%")
```
